```python
import pandas as pd
ratings = pd.read_csv("ratings.csv")
movies = pd.read_csv("movies.csv")
```

```python
# Output dataframe "ratings"
print(ratings)
```

```python
# Output dataframe "movies"
print(movies)
```

```python
# Merge the two dataframe horizontally
moviesRatings = pd.merge(ratings, movies)
print(moviesRatings)
```

```python
# We need to turn the dataframe into a pivot table of Excel
# The index will be "电影名", columns will be "用户id", values will be "评分"
userRatings = moviesRatings.pivot_table(index = "电影名", columns = "用户id", values = "评分")
userRatings
```

```python
# Solve the correlation of each user's rating pairwisely (with pearson correlation and min common movies rated of 10)
# Afterwards, put it within a n user * n user matrix (symmetric) in dataframe form
userRatingCorr = userRatings.corr(method = "pearson", min_periods = 10)
userRatingCorr
```

```python
# Turn the diagonals to Nan so that the correlation to itself (which is in turn 1) will be disregarded
import numpy as np
np.fill_diagonal(userRatingCorr.values, np.nan)
userRatingCorr
```

```python
# Create an apply function 
def findSimilarUser(x):
    return x.idxmax()

similarUsersIndex = userRatingCorr.apply(findSimilarUser)
similarUsersIndex
```

```python
# Print the correlation value of each user to the most similar user
similarUsersCorr = dict()
for i in range(6040):
    similarUsersTuple = (similarUsersIndex.index[i], similarUsersIndex.values[i])
    similarUsersCorr[similarUsersTuple] = userRatingCorr.loc[similarUsersIndex.index[i], similarUsersIndex.values[i]]
print(similarUsersCorr)
```

```python
# Checkk whether the max corresponds to the result shown above
userRatingCorr.describe().loc["max"]
```

```python
# Get the column (thus movie rating) of each respective similar user
similarUsersRatings = userRatings[similarUsersIndex.values]
similarUsersRatings
```

```python
# Change all the values (ratings) of the dataframe that is not 5 to False
# Change the values (ratings) that is 5 (to be recommended) to True
similarUsersRatings5 = pd.DataFrame(similarUsersRatings.values == 5)
similarUsersRatings5
```

```python
# To see how many movies each similar user rated 5
print(similarUsersRatings5.sum(axis=0))
```

```python
# Extract the ones that are ranked 5 and the user has not watched
moviesWatched5 = (np.isnan(np.array(userRatings))) & (similarUsersRatings.values == 5)
recommendMovies = pd.DataFrame(moviesWatched5)
recommendMovies.index = userRatings.index
recommendMovies.columns = userRatings.columns
recommendMovies
```

```python
recommendMovies.to_csv("Recommend_Movies.csv")
```
